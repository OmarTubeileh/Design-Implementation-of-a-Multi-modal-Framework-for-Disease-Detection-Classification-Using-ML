```python
import json
from collections import Counter

# Fill this in with your actual path
json_path = "/content/drive/My Drive/GP2/3_After_Text_Augmentation/Text_Augmented.json"

# Your label mapping
label_mapping = {
    "Congenital & Genetic": 0,
    "Trauma & Physical Injuries": 1,
    "Vascular & Circulatory": 2,
    "Infections": 3,
    "Neoplasm - Benign & Sarcoma": 4,
    "Neoplasm - Carcinoma": 5,
    "Neoplasm - Other Malignant": 6,
    "Inflammatory & Autoimmune": 7,
    "Metabolic & Endocrine": 8,
    "Cysts & Degenerative Conditions": 9,
    "Obstruction & Structural Abnormalities": 10,
    "Miscellaneous Conditions": 11
}

# Load JSON
with open(json_path, 'r') as f:
    data = json.load(f)

# Count label occurrences
label_counter = Counter()
for item in data:
    label_str = item.get("Class/Label", None)
    if label_str in label_mapping:
        label_counter[label_mapping[label_str]] += 1

# Print counts per class
print("ðŸ“¦ Number of instances per class:")
for label_name, label_index in label_mapping.items():
    count = label_counter.get(label_index, 0)
    print(f"{label_name}: {count}")
```
